## 벡터로 만들 수 있는 모든 것: 생성(Span)

생성(Span)은 주어진 벡터들의 모든 가능한 선형 결합(Linear Combination)의 집합을 의미한다. 즉, 주어진 벡터들을 재료로 삼아 스칼라 곱(크기 조절)과 벡터 덧셈(방향 조합)을 통해 도달할 수 있는 모든 벡터(점)들의 공간을 말한다. 이는 '주어진 벡터들이 얼마나 넓은 공간을 만들어 낼 수 있는가?'에 대한 답이다.

### 생성(Span)의 기하학적 의미

- **하나의 벡터의 Span**: 원점을 지나는 직선이다. (벡터를 늘리거나 줄이는 것만 가능)
- **두 벡터의 Span**:
  - 두 벡터가 서로 다른 방향을 가리키면(선형 독립), 그들이 만드는 Span은 하나의 평면이다.
  - 두 벡터가 같은 방향을 가리키면(선형 종속), 그들이 만드는 Span은 여전히 하나의 직선에 불과하다.
- **세 벡터의 Span (3차원 공간에서)**:
  - 세 벡터가 모두 한 평면에 있지 않으면(선형 독립), 그들의 Span은 3차원 공간 전체이다.
  - 세 벡터가 모두 한 평면에 있다면(선형 종속), 그들의 Span은 하나의 평면에 머무른다.

### 파이썬으로 생성(Span) 시각화하기

2차원 공간에서 서로 다른 방향을 가리키는 두 벡터가 어떻게 2차원 평면 전체를 생성하는지 시각적으로 확인해본다. 두 벡터의 수많은 무작위 선형 결합을 점으로 찍어보면, 그 점들이 평면을 가득 채우는 것을 볼 수 있다.

```python
import numpy as np
import matplotlib.pyplot as plt

# 시각화를 위한 설정
plt.style.use('seaborn-v0_8-whitegrid')
plt.rcParams['font.family'] = 'Malgun Gothic' # 한글 폰트 설정
plt.rcParams['axes.unicode_minus'] = False # 마이너스 부호 깨짐 방지

# 2차원 공간을 생성할 두 벡터 정의
v1 = np.array([2, 1])
v2 = np.array([-1, 2])

print(f"재료 벡터 v1: {v1}")
print(f"재료 벡터 v2: {v2}")
```

#### 선형 결합과 시각화

```python
# 100개의 무작위 선형 결합 생성
np.random.seed(0) # 결과를 동일하게 보기 위한 시드 설정
num_combinations = 100
span_vectors = []
for _ in range(num_combinations):
    # -5에서 5 사이의 무작위 스칼라(계수) 생성
    c1 = np.random.uniform(-5, 5)
    c2 = np.random.uniform(-5, 5)
    
    # 선형 결합 계산
    linear_combination = c1 * v1 + c2 * v2
    span_vectors.append(linear_combination)

span_vectors = np.array(span_vectors)

# 시각화
plt.figure(figsize=(8, 8))

# 생성된 벡터들(선형 결합 결과)을 점으로 표시
plt.scatter(span_vectors[:, 0], span_vectors[:, 1], s=20, color='gray', label='v1과 v2의 선형 결합 결과')

# 원점에서 시작하는 기본 벡터들을 화살표로 표시
plt.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='red', label='벡터 v1')
plt.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='blue', label='벡터 v2')

# 그래프 설정
limit = 15
plt.xlim(-limit, limit)
plt.ylim(-limit, limit)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.title('두 벡터 v1, v2의 생성(Span)')
plt.xlabel('x축')
plt.ylabel('y축')
plt.legend()
plt.grid(True)
plt.gca().set_aspect('equal', adjustable='box') # 가로, 세로 비율을 같게 설정
plt.show()
```

위 그래프에서 회색 점들은 모두 빨간색 벡터(v1)와 파란색 벡터(v2)의 스칼라 곱과 덧셈만으로 만들어진 결과이다. 이 점들이 2차원 평면 전체에 퍼져 있는 것을 통해, 두 벡터 v1과 v2가 2차원 평면 전체를 '생성(Span)'함을 알 수 있다.

#### 요약

생성(Span)은 주어진 벡터들의 선형 결합을 통해 도달할 수 있는 모든 점들의 집합, 즉 벡터들이 만들어내는 공간의 범위를 의미한다.
