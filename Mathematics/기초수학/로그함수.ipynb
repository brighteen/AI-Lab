# Î°úÍ∑∏ Ìï®Ïàò (Logarithmic Function)

### Í∞úÎÖê ÏöîÏïΩ
Î°úÍ∑∏ Ìï®Ïàò(Logarithmic Function)Îäî ÏßÄÏàò Ìï®Ïàò(Exponential Function)Ïùò Ïó≠Ìï®ÏàòÏûÑ. Í±∞Îì≠Ï†úÍ≥±ÏóêÏÑú Î∞ëÍ≥º Í∑∏ Í≤∞Í≥ºÍ∞íÏùÑ Ïïå Îïå, ÏßÄÏàòÎ•º Íµ¨ÌïòÍ∏∞ ÏúÑÌï¥ ÏÇ¨Ïö©ÌïòÎäî Ìï®ÏàòÏûÑ. ÌÅ∞ ÏàòÎ•º ÏûëÏùÄ ÏàòÎ°ú ÎÇòÌÉÄÎÇ¥Ïñ¥ Îã§Î£®Í∏∞ ÏâΩÍ≤å ÎßåÎì§Î©∞, Îç∞Ïù¥ÌÑ∞Ïùò Ïä§ÏºÄÏùºÏùÑ Ï°∞Ï†ïÌïòÍ±∞ÎÇò Î≥ÄÌôîÏú®ÏùÑ Î∂ÑÏÑùÌïòÎäî Îì± Îã§ÏñëÌïú Î∂ÑÏïºÏóêÏÑú ÌôúÏö©Îê® (Ïòà: ÏßÄÏßÑÏùò ÏÑ∏Í∏∞Î•º ÎÇòÌÉÄÎÇ¥Îäî Î¶¨ÌûàÌÑ∞ Í∑úÎ™®, ÏÜåÎ¶¨Ïùò ÌÅ¨Í∏∞Î•º ÎÇòÌÉÄÎÇ¥Îäî Îç∞ÏãúÎ≤®(dB), Ïö©Ïï°Ïùò ÏÇ∞ÏÑ±ÎèÑÎ•º ÎÇòÌÉÄÎÇ¥Îäî pH).

### ÏàòÌïôÏ†Å Î∞∞Í≤Ω üìê
ÏßÄÏàò Ìï®Ïàò $ y = a^x $ (Îã®, $a > 0, a \neq 1$)Í∞Ä Ï£ºÏñ¥Ï°åÏùÑ Îïå, Ïù¥Ïùò Ïó≠Ìï®ÏàòÏù∏ Î°úÍ∑∏ Ìï®ÏàòÎäî Îã§ÏùåÍ≥º Í∞ôÏù¥ Ï†ïÏùòÎê®.
$$ x = \log_a(y) $$
ÏùºÎ∞òÏ†ÅÏúºÎ°ú ÏûÖÎ†•Í∞íÏùÑ $x$, Ï∂úÎ†•Í∞íÏùÑ $y$Î°ú ÌëúÍ∏∞ÌïòÎØÄÎ°ú Îã§ÏùåÍ≥º Í∞ôÏù¥ ÎÇòÌÉÄÎÉÑ.
$$ y = \log_a(x) $$
Ïó¨Í∏∞ÏÑú $a$Îäî 'Î∞ë(base)', $x$Îäî 'ÏßÑÏàò(antilogarithm)'ÎùºÍ≥† Ìï®.
- **Ï†ïÏùòÏó≠**: ÏßÑÏàò $x$Îäî Ìï≠ÏÉÅ 0Î≥¥Îã§ Ïª§Ïïº Ìï® ($x > 0$).
- **ÏπòÏó≠**: Ìï®ÏàòÍ∞í $y$Îäî Î™®Îì† Ïã§Ïàò Í∞íÏùÑ Í∞ÄÏßà Ïàò ÏûàÏùå.
- **Ï£ºÏöî ÌäπÏßï**:
  - Î∞ë $a$Ïùò Í∞íÏóê Í¥ÄÍ≥ÑÏóÜÏù¥ Ìï≠ÏÉÅ Ï†ê $(1, 0)$ÏùÑ ÏßÄÎÇ® ($\log_a(1) = 0$).
  - Ï†ê $(a, 1)$ÏùÑ ÏßÄÎÇ® ($\log_a(a) = 1$).
  - yÏ∂ïÏùÑ Ï†êÍ∑ºÏÑ†ÏúºÎ°ú Í∞ÄÏßê.

```python
# !pip install numpy
# !pip install matplotlib

import numpy as np
import matplotlib.pyplot as plt
```

### ÏòàÏ†ú Îç∞Ïù¥ÌÑ∞ Ï§ÄÎπÑ (Data Preparation)
Í∑∏ÎûòÌîÑÎ•º Í∑∏Î¶¨Í∏∞ ÏúÑÌï¥ ÏñëÏàò Î≤îÏúÑÏùò xÍ∞íÎì§ÏùÑ ÏÉùÏÑ±Ìï®. Î°úÍ∑∏ Ìï®ÏàòÎäî Îã§ÏñëÌïú Î∞ëÏùÑ Í∞ÄÏßà Ïàò ÏûàÏúºÎØÄÎ°ú, ÎåÄÌëúÏ†ÅÏù∏ Î∞ë 2, e(ÏûêÏó∞Î°úÍ∑∏), 10(ÏÉÅÏö©Î°úÍ∑∏)Ïóê ÎåÄÌïú Î°úÍ∑∏ Í∞íÏùÑ Í∞ÅÍ∞Å Í≥ÑÏÇ∞Ìï®.

```python
# 0Ïóê Í∞ÄÍπåÏö¥ ÏûëÏùÄ ÏñëÏàòÎ∂ÄÌÑ∞ ÏãúÏûëÌïòÎäî xÍ∞í Î∞∞Ïó¥ ÏÉùÏÑ± (Î°úÍ∑∏ Ìï®Ïàò Ï†ïÏùòÏó≠ x > 0)
x_values = np.linspace(0.01, 10, 400)

# Îã§ÏñëÌïú Î∞ë(base)Ïóê ÎåÄÌïú yÍ∞í Í≥ÑÏÇ∞
y_log2 = np.log2(x_values)       # Î∞ëÏù¥ 2Ïù∏ Î°úÍ∑∏ log_2
y_loge = np.log(x_values)       # Î∞ëÏù¥ eÏù∏ ÏûêÏó∞Î°úÍ∑∏ (np.logÎäî ÏûêÏó∞Î°úÍ∑∏Í∞Ä Í∏∞Î≥∏Í∞í) log_e
y_log10 = np.log10(x_values)    # Î∞ëÏù¥ 10Ïù∏ ÏÉÅÏö©Î°úÍ∑∏ log_10

print(f"x_values (shape: {x_values.shape})")
print(f"y_log2 (shape: {y_log2.shape})")
```

### ÌïµÏã¨ Î°úÏßÅ Íµ¨ÌòÑ (Code Implementation)
NumpyÏùò ÎÇ¥Ïû• Î°úÍ∑∏ Ìï®Ïàò(`log2`, `log`, `log10`)Î•º ÏÇ¨Ïö©ÌïòÏó¨ ÌäπÏ†ï Í∞íÏóê ÎåÄÌïú Î°úÍ∑∏ Í∞íÏùÑ Í≥ÑÏÇ∞ÌïòÍ≥† ÌôïÏù∏Ìï¥Î¥Ñ. Ïù¥Î•º ÌÜµÌï¥ Î∞ëÍ≥º ÏßÑÏàòÏóê Îî∞Îùº Í≤∞Í≥ºÍ∞Ä Ïñ¥ÎñªÍ≤å Îã¨ÎùºÏßÄÎäîÏßÄ ÌôïÏù∏Ìï®.

```python
# ÌäπÏ†ï Í∞íÏóê ÎåÄÌïú Î°úÍ∑∏ Í∞í Í≥ÑÏÇ∞ ÏòàÏãú
val1 = 8
log2_val1 = np.log2(val1)
print(f"log_2({val1}) = {log2_val1:.4f} (2Î•º Î™á Î≤à Í≥±Ìï¥Ïïº {val1}Ïù¥ ÎêòÎäîÍ∞Ä?)")

val2 = 100
log10_val2 = np.log10(val2)
print(f"log_10({val2}) = {log10_val2:.4f} (10ÏùÑ Î™á Î≤à Í≥±Ìï¥Ïïº {val2}Í∞Ä ÎêòÎäîÍ∞Ä?)")

# Î∞ëÍ≥º ÏßÑÏàòÍ∞Ä Í∞ôÏúºÎ©¥ Í≤∞Í≥ºÎäî 1
val3 = np.e # ÏûêÏó∞ÏÉÅÏàò e (ÏïΩ 2.718)
loge_val3 = np.log(val3)
print(f"log_e({val3:.4f}) = {loge_val3:.4f}")

# ÏßÑÏàòÍ∞Ä 1Ïù¥Î©¥ Í≤∞Í≥ºÎäî 0
val4 = 1
log_val4 = np.log10(val4)
print(f"log_10({val4}) = {log_val4:.4f}")
```

### Í≤∞Í≥º ÌôïÏù∏ Î∞è ÏãúÍ∞ÅÌôî (Results & Visualization) üìä
Îã§ÏñëÌïú Î∞ëÏùÑ Í∞ÄÏßÑ Î°úÍ∑∏ Ìï®ÏàòÎì§ÏùÑ ÌïòÎÇòÏùò Í∑∏ÎûòÌîÑÏóê Í∑∏Î†§ ÎπÑÍµêÌï®. Î∞ëÏù¥ ÌÅ¥ÏàòÎ°ù Í∑∏ÎûòÌîÑÎäî xÏ∂ïÏóê Îçî Í∞ÄÍπùÍ≤å Í∑∏Î†§ÏßÄÎ©∞, Î™®Îì† Í∑∏ÎûòÌîÑÎäî Ï†ê (1, 0)ÏùÑ ÏßÄÎÇòÎäî Í≤ÉÏùÑ ÌôïÏù∏Ìï† Ïàò ÏûàÏùå.

```python
plt.figure(figsize=(10, 6))
plt.style.use('seaborn-v0_8-whitegrid')

# Î°úÍ∑∏ Ìï®Ïàò Í∑∏ÎûòÌîÑ
plt.plot(x_values, y_log2, label='y = log$_2$(x)')
plt.plot(x_values, y_loge, label='y = log$_e$(x) (ln x)', linestyle='--')
plt.plot(x_values, y_log10, label='y = log$_{10}$(x)', linestyle=':')

# y=x ÎåÄÏπ≠ÏùÑ Î≥¥Ïó¨Ï£ºÍ∏∞ ÏúÑÌïú y=x ÏßÅÏÑ†
plt.plot(x_values, x_values, label='y = x', color='gray', alpha=0.5)

# Ï£ºÏöî ÏßÄÏ†ê Î∞è Ï†êÍ∑ºÏÑ† ÌëúÏãú
plt.axhline(0, color='black', linewidth=0.8) # xÏ∂ï
plt.axvline(0, color='black', linewidth=0.8) # yÏ∂ï
plt.axvline(1, color='red', linestyle='-.', linewidth=1, label='x=1 (y=0)') # x=1 Ï†êÏÑ†

# Í∑∏ÎûòÌîÑ ÏÑ§Ï†ï
plt.title('Logarithmic Functions with Different Bases', fontsize=15)
plt.xlabel('x')
plt.ylabel('y = log(x)')
plt.legend()
plt.xlim(-1, 10)
plt.ylim(-4, 4)
plt.show()
```

#### ÏßÄÏàò Ìï®ÏàòÏôÄÏùò Í¥ÄÍ≥Ñ ÏãúÍ∞ÅÌôî
Î°úÍ∑∏ Ìï®ÏàòÎäî ÏßÄÏàò Ìï®ÏàòÏùò Ïó≠Ìï®ÏàòÏù¥ÎØÄÎ°ú, Îëê Ìï®ÏàòÏùò Í∑∏ÎûòÌîÑÎäî $y=x$ ÏßÅÏÑ†Ïóê ÎåÄÌï¥ ÎåÄÏπ≠ÏûÑ. Î∞ëÏù¥ $e$Ïù∏ Í≤ΩÏö∞Î•º ÏòàÎ°ú Îì§Ïñ¥ ÏßÄÏàò Ìï®Ïàò $y=e^x$ ÏôÄ Î°úÍ∑∏ Ìï®Ïàò $y=\log_e(x)$Î•º Ìï®Íªò Í∑∏Î†§ ÎåÄÏπ≠ Í¥ÄÍ≥ÑÎ•º ÌôïÏù∏Ìï®.

```python
# ÏßÄÏàòÌï®ÏàòÏôÄ Î°úÍ∑∏Ìï®Ïàò ÎπÑÍµêÎ•º ÏúÑÌïú Îç∞Ïù¥ÌÑ∞
x_exp = np.linspace(-3, 3, 400)
y_exp = np.exp(x_exp) # y = e^x

x_log = np.linspace(0.05, 15, 400)
y_log = np.log(x_log) # y = log_e(x)

plt.figure(figsize=(8, 8))
plt.plot(x_exp, y_exp, label='Exponential: y = e$^x$')
plt.plot(x_log, y_log, label='Logarithmic: y = log$_e$(x)')
plt.plot([-4, 8], [-4, 8], 'k--', alpha=0.7, label='y = x (Line of Symmetry)')

plt.title('Inverse Relationship between Exponential and Logarithmic Functions', fontsize=14)
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.xlim(-4, 8)
plt.ylim(-4, 8)
plt.gca().set_aspect('equal', adjustable='box') # Í∞ÄÎ°ú, ÏÑ∏Î°ú ÎπÑÏú®ÏùÑ Í∞ôÍ≤å ÏÑ§Ï†ï
plt.show()
```

### Í≥†Î†§ÏÇ¨Ìï≠ (Considerations)
- <b>Ï†ïÏùòÏó≠ Ïò§Î•ò(Domain Error)</b>: Î°úÍ∑∏ Ìï®ÏàòÏùò ÏßÑÏàò(ÏûÖÎ†•Í∞í)Îäî Ìï≠ÏÉÅ ÏñëÏàòÏó¨Ïïº Ìï®. ÏΩîÎìúÏóêÏÑú 0 ÎòêÎäî ÏùåÏàòÎ•º ÏûÖÎ†•ÌïòÎ©¥ `nan` (Not a Number) Í∞íÏù¥ Î∞òÌôòÎêòÍ±∞ÎÇò Ïò§Î•òÍ∞Ä Î∞úÏÉùÌï®.
- **Î∞ëÏùò Ï°∞Í±¥**: Î°úÍ∑∏Ïùò Î∞ë $a$Îäî 1Ïù¥ ÏïÑÎãå ÏñëÏàò($a>0, a \neq 1$)Ïó¨Ïïº Ìï®. ÎßåÏïΩ Î∞ëÏù¥ 1Ïù¥ÎùºÎ©¥ $\log_1(x)$Îäî xÍ∞Ä 1Ïùº ÎïåÎ•º Ï†úÏô∏ÌïòÍ≥†Îäî Ï†ïÏùòÎêòÏßÄ ÏïäÏùå.
- **Î∞ëÏùò ÌÅ¨Í∏∞Ïóê Îî∞Î•∏ Î≥ÄÌôî**: Î∞ë $a$Í∞Ä 1Î≥¥Îã§ ÌÅ¨Î©¥ Ìï®ÏàòÎäî Ï¶ùÍ∞ÄÌïòÍ≥†, $0 < a < 1$Ïù¥Î©¥ Ìï®ÏàòÎäî Í∞êÏÜåÌï®. Î∞ëÏù¥ ÌÅ¥ÏàòÎ°ù Í∑∏ÎûòÌîÑÎäî xÏ∂ïÏóê Îçî Í∞ÄÍπùÍ≤å, ÏôÑÎßåÌïòÍ≤å Ï¶ùÍ∞ÄÌï®.
