# 로그 함수 (Logarithmic Function)

### 개념 요약
로그 함수(Logarithmic Function)는 지수 함수(Exponential Function)의 역함수임. 거듭제곱에서 밑과 그 결과값을 알 때, 지수를 구하기 위해 사용하는 함수임. 큰 수를 작은 수로 나타내어 다루기 쉽게 만들며, 데이터의 스케일을 조정하거나 변화율을 분석하는 등 다양한 분야에서 활용됨 (예: 지진의 세기를 나타내는 리히터 규모, 소리의 크기를 나타내는 데시벨(dB), 용액의 산성도를 나타내는 pH).

### 수학적 배경 📐
지수 함수 $ y = a^x $ (단, $a > 0, a \neq 1$)가 주어졌을 때, 이의 역함수인 로그 함수는 다음과 같이 정의됨.
$$ x = \log_a(y) $$
일반적으로 입력값을 $x$, 출력값을 $y$로 표기하므로 다음과 같이 나타냄.
$$ y = \log_a(x) $$
여기서 $a$는 '밑(base)', $x$는 '진수(antilogarithm)'라고 함.
- **정의역**: 진수 $x$는 항상 0보다 커야 함 ($x > 0$).
- **치역**: 함수값 $y$는 모든 실수 값을 가질 수 있음.
- **주요 특징**:
  - 밑 $a$의 값에 관계없이 항상 점 $(1, 0)$을 지남 ($\log_a(1) = 0$).
  - 점 $(a, 1)$을 지남 ($\log_a(a) = 1$).
  - y축을 점근선으로 가짐.

```python
# !pip install numpy
# !pip install matplotlib

import numpy as np
import matplotlib.pyplot as plt
```

### 예제 데이터 준비 (Data Preparation)
그래프를 그리기 위해 양수 범위의 x값들을 생성함. 로그 함수는 다양한 밑을 가질 수 있으므로, 대표적인 밑 2, e(자연로그), 10(상용로그)에 대한 로그 값을 각각 계산함.

```python
# 0에 가까운 작은 양수부터 시작하는 x값 배열 생성 (로그 함수 정의역 x > 0)
x_values = np.linspace(0.01, 10, 400)

# 다양한 밑(base)에 대한 y값 계산
y_log2 = np.log2(x_values)       # 밑이 2인 로그 log_2
y_loge = np.log(x_values)       # 밑이 e인 자연로그 (np.log는 자연로그가 기본값) log_e
y_log10 = np.log10(x_values)    # 밑이 10인 상용로그 log_10

print(f"x_values (shape: {x_values.shape})")
print(f"y_log2 (shape: {y_log2.shape})")
```

### 핵심 로직 구현 (Code Implementation)
Numpy의 내장 로그 함수(`log2`, `log`, `log10`)를 사용하여 특정 값에 대한 로그 값을 계산하고 확인해봄. 이를 통해 밑과 진수에 따라 결과가 어떻게 달라지는지 확인함.

```python
# 특정 값에 대한 로그 값 계산 예시
val1 = 8
log2_val1 = np.log2(val1)
print(f"log_2({val1}) = {log2_val1:.4f} (2를 몇 번 곱해야 {val1}이 되는가?)")

val2 = 100
log10_val2 = np.log10(val2)
print(f"log_10({val2}) = {log10_val2:.4f} (10을 몇 번 곱해야 {val2}가 되는가?)")

# 밑과 진수가 같으면 결과는 1
val3 = np.e # 자연상수 e (약 2.718)
loge_val3 = np.log(val3)
print(f"log_e({val3:.4f}) = {loge_val3:.4f}")

# 진수가 1이면 결과는 0
val4 = 1
log_val4 = np.log10(val4)
print(f"log_10({val4}) = {log_val4:.4f}")
```

### 결과 확인 및 시각화 (Results & Visualization) 📊
다양한 밑을 가진 로그 함수들을 하나의 그래프에 그려 비교함. 밑이 클수록 그래프는 x축에 더 가깝게 그려지며, 모든 그래프는 점 (1, 0)을 지나는 것을 확인할 수 있음.

```python
plt.figure(figsize=(10, 6))
plt.style.use('seaborn-v0_8-whitegrid')

# 로그 함수 그래프
plt.plot(x_values, y_log2, label='y = log$_2$(x)')
plt.plot(x_values, y_loge, label='y = log$_e$(x) (ln x)', linestyle='--')
plt.plot(x_values, y_log10, label='y = log$_{10}$(x)', linestyle=':')

# y=x 대칭을 보여주기 위한 y=x 직선
plt.plot(x_values, x_values, label='y = x', color='gray', alpha=0.5)

# 주요 지점 및 점근선 표시
plt.axhline(0, color='black', linewidth=0.8) # x축
plt.axvline(0, color='black', linewidth=0.8) # y축
plt.axvline(1, color='red', linestyle='-.', linewidth=1, label='x=1 (y=0)') # x=1 점선

# 그래프 설정
plt.title('Logarithmic Functions with Different Bases', fontsize=15)
plt.xlabel('x')
plt.ylabel('y = log(x)')
plt.legend()
plt.xlim(-1, 10)
plt.ylim(-4, 4)
plt.show()
```

#### 지수 함수와의 관계 시각화
로그 함수는 지수 함수의 역함수이므로, 두 함수의 그래프는 $y=x$ 직선에 대해 대칭임. 밑이 $e$인 경우를 예로 들어 지수 함수 $y=e^x$ 와 로그 함수 $y=\log_e(x)$를 함께 그려 대칭 관계를 확인함.

```python
# 지수함수와 로그함수 비교를 위한 데이터
x_exp = np.linspace(-3, 3, 400)
y_exp = np.exp(x_exp) # y = e^x

x_log = np.linspace(0.05, 15, 400)
y_log = np.log(x_log) # y = log_e(x)

plt.figure(figsize=(8, 8))
plt.plot(x_exp, y_exp, label='Exponential: y = e$^x$')
plt.plot(x_log, y_log, label='Logarithmic: y = log$_e$(x)')
plt.plot([-4, 8], [-4, 8], 'k--', alpha=0.7, label='y = x (Line of Symmetry)')

plt.title('Inverse Relationship between Exponential and Logarithmic Functions', fontsize=14)
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.xlim(-4, 8)
plt.ylim(-4, 8)
plt.gca().set_aspect('equal', adjustable='box') # 가로, 세로 비율을 같게 설정
plt.show()
```

### 고려사항 (Considerations)
- <b>정의역 오류(Domain Error)</b>: 로그 함수의 진수(입력값)는 항상 양수여야 함. 코드에서 0 또는 음수를 입력하면 `nan` (Not a Number) 값이 반환되거나 오류가 발생함.
- **밑의 조건**: 로그의 밑 $a$는 1이 아닌 양수($a>0, a \neq 1$)여야 함. 만약 밑이 1이라면 $\log_1(x)$는 x가 1일 때를 제외하고는 정의되지 않음.
- **밑의 크기에 따른 변화**: 밑 $a$가 1보다 크면 함수는 증가하고, $0 < a < 1$이면 함수는 감소함. 밑이 클수록 그래프는 x축에 더 가깝게, 완만하게 증가함.
