# 지수 함수 (Exponential Function)

### 개념 요약
**지수 함수**는 거듭제곱의 지수에 변수 $x$를 포함하는 함수로, 어떤 양이 특정 비율로 지속해서 증가하거나 감소하는 현상을 설명하는 데 사용됨. 이 함수는 자연 현상, 금융, 공학 등 다양한 분야에서 중요한 역할을 하며, 특히 복리 계산, 인구 증가, 방사성 붕괴 모델링에 핵심적으로 활용됨.

### 수학적 배경 📐
지수 함수는 일반적으로 다음과 같은 형태로 표현됨:
$$ f(x) = a^x $$
여기서 `a`는 **밑(base)**이라 불리는 0보다 큰 상수이며, 1이 아님. `x`는 **지수(exponent)**라 불리는 변수임.
- 밑 `a`가 1보다 크면 ($a > 1$), 함수는 **지수적 증가(exponential growth)**를 보임.
- 밑 `a`가 0과 1 사이에 있으면 ($0 < a < 1$), 함수는 **지수적 감소(exponential decay)**를 보임.

### 라이브러리 임포트 (Setup)

```python
# !pip install numpy
# !pip install matplotlib

import numpy as np
import matplotlib.pyplot as plt

# 그래프 스타일 설정
plt.style.use('seaborn-v0_8-whitegrid')
```

### 예제 데이터 준비 (Data Preparation)
지수 함수 그래프를 그리기 위해 x값의 범위를 정의함.

```python
# -2부터 2까지 100개의 점으로 x 데이터 생성
x = np.linspace(-2, 2, 100)

print(f"x 데이터의 형태 (shape): {x.shape}")
print(f"x 데이터 (앞 5개): {x[:5]}")
```

### 핵심 로직 구현 (Code Implementation)
다양한 밑(base)을 가진 지수 함수를 정의하고, 각 x값에 대한 y값을 계산함.

```python
# 밑(base)이 2인 경우 (지수적 증가)
base_2 = 2
y_growth = base_2**x

# 밑(base)이 자연상수 e (약 2.718)인 경우
base_e = np.e
y_natural_growth = base_e**x

# 밑(base)이 0.5인 경우 (지수적 감소)
base_0_5 = 0.5
y_decay = base_0_5**x

print(f"밑이 {base_2}일 때 y값 (앞 5개): {y_growth[:5]}")
print(f"밑이 e일 때 y값 (앞 5개): {y_natural_growth[:5]}")
print(f"밑이 {base_0_5}일 때 y값 (앞 5개): {y_decay[:5]}")
```

### 결과 확인 및 시각화 (Results & Visualization) 📊
계산된 결과를 바탕으로 다양한 밑을 가진 지수 함수 그래프를 시각화하여 비교함.

```python
plt.figure(figsize=(10, 6))

# 지수 함수 그래프
plt.plot(x, y_growth, label='$y=2^x$', color='blue')
plt.plot(x, y_natural_growth, label=f'$y=e^x \\approx {np.e:.2f}^x$', color='green', linestyle='--')
plt.plot(x, y_decay, label='$y=0.5^x$', color='red')

# 그래프 제목 및 축 레이블 설정
plt.title('Exponential Functions with Different Bases', fontsize=15)
plt.xlabel('x', fontsize=12)
plt.ylabel('$f(x) = a^x$', fontsize=12)
plt.legend(fontsize=12)

# y축과 x축 교차점 표시
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.ylim(-1, 6)
plt.grid(True)

plt.show()
```

### 고려사항 (Considerations)
- **밑(base)의 중요성**: 밑 `a`의 값에 따라 그래프의 형태가 급격히 변함. `a`가 1에 가까울수록 변화가 완만하고, 1에서 멀어질수록 변화가 가파르게 나타남.
- **정의역과 치역**: 지수 함수 $f(x)=a^x$의 정의역(x값의 범위)은 모든 실수이지만, 치역(y값의 범위)은 항상 0보다 큼. 그래프는 x축에 닿지 않음 (x축이 점근선).
- **자연상수 e**: 과학과 공학 분야에서는 밑으로 자연상수 $e \approx 2.71828$를 사용하는 경우가 매우 많음. 이를 자연로그의 밑이라고 하며, 관련 함수를 미분하거나 적분할 때 계산이 편리해지는 특징이 있음.
