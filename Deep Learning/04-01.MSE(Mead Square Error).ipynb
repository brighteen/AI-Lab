# í‰ê·  ì œê³± ì˜¤ì°¨ (Mean Squared Error, MSE)

**í‰ê·  ì œê³± ì˜¤ì°¨(MSE)**ëŠ” ì˜ˆì¸¡ê°’ê³¼ ì‹¤ì œê°’ ì‚¬ì´ì˜ ì°¨ì´(ì˜¤ì°¨)ë¥¼ ì œê³±í•˜ì—¬ í‰ê· ì„ ë‚¸ ê°’ì„. íšŒê·€ ë¬¸ì œì—ì„œ ëª¨ë¸ì˜ ì„±ëŠ¥ì„ ì¸¡ì •í•˜ëŠ” ëŒ€í‘œì ì¸ ì†ì‹¤ í•¨ìˆ˜(Loss Function) ë˜ëŠ” í‰ê°€ì§€í‘œë¡œ ì‚¬ìš©ë¨.

MSEëŠ” ì˜¤ì°¨ì˜ í¬ê¸°ì— ë¯¼ê°í•˜ë©°, ê°’ì´ ì‘ì„ìˆ˜ë¡ ëª¨ë¸ì˜ ì˜ˆì¸¡ì´ ì‹¤ì œê°’ì— ê°€ê¹ë‹¤ëŠ” ê²ƒì„ ì˜ë¯¸í•¨. ëª¨ë¸ í•™ìŠµ ê³¼ì •ì—ì„œ ì´ MSE ê°’ì„ ìµœì†Œí™”í•˜ëŠ” ë°©í–¥ìœ¼ë¡œ ê°€ì¤‘ì¹˜ë¥¼ ì—…ë°ì´íŠ¸í•¨.

### ğŸ“ ìˆ˜í•™ì  ë°°ê²½

nê°œì˜ ë°ì´í„°ì— ëŒ€í•´ ì‹¤ì œê°’ $y$ì™€ ëª¨ë¸ì˜ ì˜ˆì¸¡ê°’ $\hat{y}$ê°€ ìˆì„ ë•Œ, MSEëŠ” ë‹¤ìŒê³¼ ê°™ì´ ì •ì˜ë¨.

$$ MSE = \frac{1}{n} \sum_{i=1}^{n}(y_i - \hat{y}_i)^2 $$

* $n$: ì „ì²´ ë°ì´í„°ì˜ ê°œìˆ˜
* $y_i$: ië²ˆì§¸ ë°ì´í„°ì˜ ì‹¤ì œê°’
* $\hat{y}_i$: ië²ˆì§¸ ë°ì´í„°ì— ëŒ€í•œ ëª¨ë¸ì˜ ì˜ˆì¸¡ê°’

### ì…‹ì—… (Setup)
ê³„ì‚° ë° ì‹œê°í™”ë¥¼ ìœ„í•œ ë¼ì´ë¸ŒëŸ¬ë¦¬ ì„í¬íŠ¸.

```python
# !pip install numpy matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

### ì˜ˆì œ ë°ì´í„° ì¤€ë¹„ (Data Preparation)
ê°œë… ì„¤ëª…ì„ ìœ„í•œ ê°„ë‹¨í•œ ì‹¤ì œê°’(y_true)ê³¼ ì˜ˆì¸¡ê°’(y_pred)ì„ ìƒì„±í•¨.

```python
# ì‹¤ì œê°’: y = 2x + 1 í˜•íƒœì˜ ì„ í˜• ë°ì´í„°
X = np.array([1, 2, 3, 4, 5])
y_true = np.array([3, 5, 7, 9, 11])

# ëª¨ë¸ì´ ì˜ˆì¸¡í–ˆë‹¤ê³  ê°€ì •í•˜ëŠ” ê°’ (ì•½ê°„ì˜ ì˜¤ì°¨ í¬í•¨)
y_pred = np.array([2.5, 5.2, 8.0, 8.5, 11.3])

print(f"ì…ë ¥ ë°ì´í„°(X): {X}")
print(f"ì‹¤ì œê°’(y_true): {y_true}")
print(f"ì˜ˆì¸¡ê°’(y_pred): {y_pred}")
```

### í•µì‹¬ ë¡œì§ êµ¬í˜„ (Code Implementation)
MSE ê³„ì‚° ê³¼ì •ì„ ë‹¨ê³„ë³„ë¡œ ë‚˜ëˆ„ì–´ êµ¬í˜„í•¨.

```python
# Step 1: ê° ë°ì´í„° í¬ì¸íŠ¸ì˜ ì˜¤ì°¨(error) ê³„ì‚° (y_true - y_pred)
errors = y_true - y_pred

print(f"ì˜¤ì°¨ (errors): shape={errors.shape}\n{errors}")
```

```python
# Step 2: ì˜¤ì°¨ë¥¼ ì œê³± (squared errors)
# ì´ìœ : ì˜¤ì°¨ì˜ ë¶€í˜¸(+, -)ë¥¼ ì—†ì• ê³ , í° ì˜¤ì°¨ì— ë” í° íŒ¨ë„í‹°ë¥¼ ë¶€ì—¬í•˜ê¸° ìœ„í•¨.
squared_errors = errors ** 2

print(f"ì œê³±ëœ ì˜¤ì°¨ (squared_errors): shape={squared_errors.shape}\n{squared_errors}")
```

```python
# Step 3: ì œê³±ëœ ì˜¤ì°¨ì˜ í‰ê· (mean) ê³„ì‚°
mse_value = np.mean(squared_errors)

print(f"ìµœì¢… MSE ê°’: {mse_value:.4f}")
```

#### í•¨ìˆ˜ë¡œ ì „ì²´ ê³¼ì • í†µí•©
ìœ„ ê³¼ì •ì„ í•˜ë‚˜ì˜ í•¨ìˆ˜ë¡œ ë§Œë“¤ì–´ ì‚¬ìš©.

```python
def mean_squared_error(y_true, y_pred):
    """Numpyë¥¼ ì‚¬ìš©í•˜ì—¬ MSEë¥¼ ê³„ì‚°í•˜ëŠ” í•¨ìˆ˜"""
    # ë¡œì§ì˜ ëª…í™•ì„±ì„ ìœ„í•´ í•œ ì¤„ë¡œ í‘œí˜„
    return np.mean((y_true - y_pred) ** 2)

# í•¨ìˆ˜ë¥¼ ì‚¬ìš©í•˜ì—¬ MSE ê³„ì‚°
mse_by_function = mean_squared_error(y_true, y_pred)

print(f"í•¨ìˆ˜ë¡œ ê³„ì‚°í•œ MSE: {mse_by_function:.4f}")
```

### ğŸ“Š ê²°ê³¼ í™•ì¸ ë° ì‹œê°í™” (Results & Visualization)
ì‹¤ì œê°’ê³¼ ì˜ˆì¸¡ê°’, ê·¸ë¦¬ê³  ê·¸ ì‚¬ì´ì˜ ì˜¤ì°¨ë¥¼ ì‹œê°í™”í•˜ì—¬ MSEì˜ ì˜ë¯¸ë¥¼ ì§ê´€ì ìœ¼ë¡œ ì´í•´í•¨.

```python
plt.figure(figsize=(8, 6))

# ì‹¤ì œê°’(íŒŒë€ìƒ‰ ì )ê³¼ ì˜ˆì¸¡ê°’(ë¹¨ê°„ìƒ‰ ì„ )ì„ ê·¸ë˜í”„ì— í‘œì‹œ
plt.scatter(X, y_true, color='blue', label='Actual Values ($y_{true}$)', s=100) # sëŠ” ì ì˜ í¬ê¸°
plt.plot(X, y_pred, color='red', marker='x', linestyle='--', label='Predicted Values ($\hat{y}_{pred}$)')

# ê° ë°ì´í„° í¬ì¸íŠ¸ì˜ ì˜¤ì°¨ë¥¼ íšŒìƒ‰ ì ì„ ìœ¼ë¡œ í‘œì‹œ
for i in range(len(X)):
    plt.vlines(X[i], ymin=y_pred[i], ymax=y_true[i], color='gray', linestyle=':')

plt.title('Mean Squared Error (MSE) Visualization')
plt.xlabel('Input (X)')
plt.ylabel('Value (Y)')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

### ê³ ë ¤ì‚¬í•­ (Considerations)

* **ì´ìƒì¹˜(Outlier)ì— ëŒ€í•œ ë¯¼ê°ë„**: ì˜¤ì°¨ë¥¼ ì œê³±í•˜ê¸° ë•Œë¬¸ì—, ë‹¤ë¥¸ ê°’ë“¤ì— ë¹„í•´ ìœ ë‚œíˆ í° ì˜¤ì°¨(ì´ìƒì¹˜)ê°€ ë°œìƒí•˜ë©´ MSE ê°’ì´ í¬ê²Œ ì¦ê°€í•¨. ì´ëŠ” ëª¨ë¸ì´ ì´ìƒì¹˜ì— ê³¼ë„í•˜ê²Œ ì˜í–¥ì„ ë°›ë„ë¡ ë§Œë“¤ ìˆ˜ ìˆìŒ.
* **ë‹¨ìœ„ í•´ì„ì˜ ì–´ë ¤ì›€**: MSEëŠ” ì‹¤ì œê°’ì˜ ë‹¨ìœ„ë¥¼ ì œê³±í•œ í˜•íƒœ(ì˜ˆ: ê°€ê²©ì„ ì˜ˆì¸¡í•˜ëŠ” ê²½ìš° 'ì›Â²')ê°€ ë˜ë¯€ë¡œ ì§ê´€ì ì¸ í•´ì„ì´ ì–´ë ¤ìš¸ ìˆ˜ ìˆìŒ. ì´ëŸ° ê²½ìš° MSEì— ì œê³±ê·¼ì„ ì”Œìš´ **RMSE(Root Mean Squared Error)**ë¥¼ ì‚¬ìš©í•˜ë©´ ë‹¨ìœ„ê°€ ì¼ì¹˜í•˜ì—¬ í•´ì„ì´ ìš©ì´í•´ì§.
