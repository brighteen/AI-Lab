# λ¨λ©ν…€ (Momentum)

### κ°λ… μ”μ•½
λ¨λ©ν…€μ€ κ²½μ‚¬ν•κ°•λ²•(Gradient Descent)μ„ κ°μ„ ν•λ” μµμ ν™” μ•κ³ λ¦¬μ¦ μ¤‘ ν•λ‚μ„. μ΄μ „ κ²½μ‚¬(gradient) κ°’λ“¤μ„ λ„μ ν•μ—¬ ν„μ¬ νλΌλ―Έν„° μ—…λ°μ΄νΈμ— λ°μν•λ” λ°©μ‹μΌλ΅, λ§μΉ μ–Έλ•μ„ λ‚΄λ ¤μ¤λ” κ³µμ΄ κ΄€μ„±μ„ λ°›μ•„ λ” λΉ λ¥΄κ³  μ•μ •μ μΌλ΅ μ›€μ§μ΄λ” κ²ƒκ³Ό μ μ‚¬ν•¨.

μ΄λ¬ν• κ΄€μ„± λ•λ¶„μ— μ†μ‹¤ ν•¨μ(loss function)μ ν‘λ©΄μ΄ μ™„λ§ν•κ±°λ‚ μ§€μ—­ μµμ†μ (local minima)μ΄ μλ” κ²½μ°μ—λ„ λ” ν¨κ³Όμ μΌλ΅ μ „μ—­ μµμ†μ (global minimum)μ„ μ°Ύμ•„κ° μ μμ.

### π“ μν•™μ  λ°°κ²½
λ¨λ©ν…€μ μ—…λ°μ΄νΈ κ·μΉ™μ€ λ‹¤μκ³Ό κ°™μ΄ μ •μλ¨. μ΄ κ³Όμ •μ€ 'μ†λ„(velocity)'λΌλ” κ°λ…μ„ λ„μ…ν•μ—¬ μ΄μ „ μ—…λ°μ΄νΈμ λ°©ν–¥κ³Ό ν¬κΈ°λ¥Ό κΈ°μ–µν•¨.

1. **μ†λ„(Velocity) μ—…λ°μ΄νΈ:**
$$ v_t = \gamma v_{t-1} + \eta \nabla J(W) $$

2. **νλΌλ―Έν„°(Weight) μ—…λ°μ΄νΈ:**
$$ W = W - v_t $$

* $v_t$: ν„μ¬ νƒ€μ„μ¤ν…(t)μ—μ„μ μ†λ„ (μ—…λ°μ΄νΈν•  λ°©ν–¥κ³Ό ν¬κΈ°)
* $\gamma$: λ¨λ©ν…€ κ³„μ(momentum coefficient). μ΄μ „ μ†λ„λ¥Ό μ–Όλ§λ‚ λ°μν• μ§€ κ²°μ •ν•λ” κ°’ (λ³΄ν†µ 0.9μ— κ°€κΉμ΄ κ°’ μ‚¬μ©)
* $v_{t-1}$: μ΄μ „ νƒ€μ„μ¤ν…(t-1)μ—μ„μ μ†λ„
* $\eta$: ν•™μµλ¥ (learning rate)
* $\nabla J(W)$: ν„μ¬ νλΌλ―Έν„° $W$μ— λ€ν• λΉ„μ© ν•¨μ $J$μ κ²½μ‚¬(gradient)

### λΌμ΄λΈλ¬λ¦¬ μ„ν¬νΈ (Setup)

```python
# !pip install numpy matplotlib

import numpy as np
import matplotlib.pyplot as plt
```

### μμ  λ°μ΄ν„° μ¤€λΉ„ (Data Preparation)
λ¨λ©ν…€μ ν¨κ³Όλ¥Ό μ‹κ°μ μΌλ΅ ν™•μΈν•κΈ° μ„ν•΄ κ°„λ‹¨ν• 1μ°¨μ› 2μ°¨ ν•¨μ($f(x) = x^2$)λ¥Ό μ†μ‹¤ ν•¨μλ΅ κ°€μ •ν•¨. μ΄ ν•¨μμ κ²½μ‚¬λ” $f'(x) = 2x$μ„.

```python
# μ†μ‹¤ ν•¨μ μ •μ (f(x) = x^2)
def loss_function(x):
    return x**2

# μ†μ‹¤ ν•¨μμ κ²½μ‚¬(λ―Έλ¶„) μ •μ (f'(x) = 2x)
def gradient(x):
    return 2*x
```

### ν•µμ‹¬ λ΅μ§ κµ¬ν„ (Code Implementation)
ν‘μ¤€ κ²½μ‚¬ν•κ°•λ²•κ³Ό λ¨λ©ν…€μ„ μ μ©ν• κ²½μ‚¬ν•κ°•λ²•μ„ κ°κ° κµ¬ν„ν•μ—¬ νλΌλ―Έν„°κ°€ μ–΄λ–»κ² μ—…λ°μ΄νΈλλ”μ§€ λΉ„κµν•¨.

```python
# ν•μ΄νΌνλΌλ―Έν„° μ„¤μ •
learning_rate = 0.1 # ν•™μµλ¥ 
gamma = 0.9         # λ¨λ©ν…€ κ³„μ
n_iterations = 20   # λ°λ³µ νμ
start_x = 10.0      # μ‹μ‘ νλΌλ―Έν„° κ°’

# κ²°κ³Όλ¥Ό μ €μ¥ν•  λ¦¬μ¤νΈ μ΄κΈ°ν™”
gd_history = []
momentum_history = []

### 1. ν‘μ¤€ κ²½μ‚¬ν•κ°•λ²• (Standard Gradient Descent)
x_gd = start_x
for i in range(n_iterations):
    gd_history.append(x_gd)
    grad = gradient(x_gd) # κ²½μ‚¬ κ³„μ‚°
    x_gd = x_gd - learning_rate * grad # νλΌλ―Έν„° μ—…λ°μ΄νΈ

### 2. λ¨λ©ν…€ (Momentum)
x_momentum = start_x
velocity = 0.0 # μ†λ„ μ΄κΈ°ν™”

print("--- λ¨λ©ν…€ μµμ ν™” κ³Όμ • ---")
for i in range(n_iterations):
    momentum_history.append(x_momentum)
    grad = gradient(x_momentum)
    
    # μ΄μ „ μ†λ„μ™€ ν„μ¬ κ²½μ‚¬λ¥Ό μ‚¬μ©ν•μ—¬ μƒλ΅μ΄ μ†λ„λ¥Ό κ³„μ‚°
    # μ΄κ²ƒμ΄ λ¨λ©ν…€μ ν•µμ‹¬ μ•„μ΄λ””μ–΄μ„
    velocity = gamma * velocity + learning_rate * grad
    
    # μƒλ΅μ΄ μ†λ„λ¥Ό μ‚¬μ©ν•μ—¬ νλΌλ―Έν„° μ—…λ°μ΄νΈ
    x_momentum = x_momentum - velocity

    if i % 2 == 0: # 2λ²μ λ°λ³µλ§λ‹¤ μ¤‘κ°„ κ³Όμ • μ¶λ ¥
        print(f"Iter {i+1:2d}: Grad={grad:8.4f}, Velocity={velocity:8.4f}, New_X={x_momentum:8.4f}")
```

### π“ κ²°κ³Ό ν™•μΈ λ° μ‹κ°ν™” (Results & Visualization)
λ‘ μµμ ν™” λ°©λ²•μ νλΌλ―Έν„° μ—…λ°μ΄νΈ κ²½λ΅λ¥Ό μ‹κ°ν™”ν•μ—¬ λΉ„κµν•¨. λ¨λ©ν…€μ΄ λ” μ μ€ λ°λ³µμΌλ΅, λ” λ¶€λ“λ½κ² μµμ μ μ— μλ ΄ν•λ” κ²ƒμ„ ν™•μΈν•  μ μμ.

```python
# μ†μ‹¤ ν•¨μ κ·Έλν”„ μƒμ„±
x_vals = np.linspace(-12, 12, 100)
y_vals = loss_function(x_vals)

plt.figure(figsize=(12, 8))
plt.plot(x_vals, y_vals, label='Loss Function ($x^2$)', color='gray')

# ν‘μ¤€ κ²½μ‚¬ν•κ°•λ²• κ²½λ΅ μ‹κ°ν™”
plt.plot(gd_history, loss_function(np.array(gd_history)), 'o--', label='Standard Gradient Descent', markersize=5)

# λ¨λ©ν…€ κ²½λ΅ μ‹κ°ν™”
plt.plot(momentum_history, loss_function(np.array(momentum_history)), 'o-', label='Momentum', markersize=8)

plt.title('Gradient Descent vs Momentum', fontsize=16)
plt.xlabel('Parameter (x)')
plt.ylabel('Loss f(x)')
plt.legend()
plt.grid(True)
plt.show()
```

### κ³ λ ¤μ‚¬ν•­ (Considerations)

* <b>λ¨λ©ν…€ κ³„μ($\gamma$)μ μ—­ν• </b>: `gamma` κ°’μ΄ ν΄μλ΅(μ: 0.99) μ΄μ „ μ—…λ°μ΄νΈμ μν–¥μ„ λ” λ§μ΄ λ°›μ•„ κ΄€μ„±μ΄ μ»¤μ§. μ΄λ΅ μΈν•΄ μµμ μ μ„ μ§€λ‚μΉ  μλ„ μμ. λ°λ©΄, κ°’μ΄ μ‘μΌλ©΄(μ: 0.5) ν‘μ¤€ κ²½μ‚¬ν•κ°•λ²•κ³Ό μ μ‚¬ν•κ² λ™μ‘ν•¨.
